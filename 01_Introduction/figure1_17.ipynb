```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Declare variables values
x1=[-1,-0.5,1, 2, 2, 1, 0.5, -1]
x2 = [0]*16;
x2[:8]=x1;

# Compute FFT
X1 = np.fft.fft(x1)
X2 = np.fft.fft(x2)
```

```python
# Ploting
plt.rcParams['figure.figsize'] = (14, 10)# figure size

plt.subplot(221)
plt.stem(x1)
plt.ylabel('x1[n]')
plt.title('8 samples')
plt.axis([-0.5, 7.5, -1.5, 2.5])

plt.subplot(222)
plt.stem(np.abs(X1))
plt.ylabel('|X1[k]|')
plt.title('8-point FFT')
plt.axis([-0.5, 7.5, -0.5, 8])

plt.subplot(223)
plt.stem(x2)
plt.ylabel('x2[n]')
plt.xlabel('n')
plt.title('8 samples + zero-padding')
plt.axis([-0.5, 15.5, -1.5, 2.5])

plt.subplot(224)
plt.stem(np.abs(X2))
plt.ylabel('|X2[k]|')
plt.xlabel('k')
plt.title('16-point FFT')
plt.axis([-1, 16, -0.5, 10])

plt.suptitle('Zero-padding to increase frequency resolution',size=16)

plt.tight_layout()
plt.subplots_adjust(top=0.9)
plt.show()
```
