```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.io import wavfile
import IPython
from IPython.display import Audio
```

```python
AudioFileName = "ton2.wav"
IPython.display.Audio(AudioFileName)
```

```python
# Load audio File
fs, data = wavfile.read(AudioFileName)
print(AudioFileName+', Sample Rate = '+str(fs))
```

```python
# Plot audio signal
plt.rcParams['figure.figsize'] = (12, 5)# figure size

plt.subplot(311)
plt.plot(data[:8000])
plt.ylabel('x (n)')

plt.subplot(312)
plt.plot(data[:1000])
plt.ylabel('x (n)')

plt.subplot(313)
plt.stem(data[:100])
plt.ylabel('x [n]')

plt.tight_layout()
plt.show()
```
