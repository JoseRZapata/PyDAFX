```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
from scipy.fftpack import fft
```

```python
x=[-1, -0.5, 1, 2, 2, 1, 0.5, -1]
w = signal.blackman(8)
w=w*8/float(sum(w))
x1=x*w
x2=[0]*16
x2[:8]=x1
```

```python
# Ploting Signals
plt.rcParams['figure.figsize'] = (10, 8)# figure size

plt.subplot(421)
plt.stem(x);plt.axis([-0.5, 7.5, -1.5, 2.5]);
plt.ylabel('x(n)');
plt.title('a) 8 samples');
plt.subplot(423);
plt.stem(w);plt.axis([-0.5, 7.5, -1.5, 3]);
plt.ylabel('w(n)');
plt.title('b) 8 samples Blackman window');

plt.subplot(425);
plt.stem(x1);plt.axis([-0.5, 7.5, -1.5, 6]);
plt.ylabel('x_w(n)');
plt.title('c) x(n)* w(n)');

plt.subplot(427);
plt.stem(x2);plt.axis([-0.5, 15.5, -1.5, 6]);
plt.xlabel('n');plt.ylabel('x_w(n)');
plt.title('d) x(n)*w(n) + zero-padding');

plt.subplot(222);
plt.stem(abs(fft(x1)));plt.axis([-0.5, 7.5, -0.5, 15]);
plt.ylabel('|X(k)|');
plt.title('8-point FFT of c)');

plt.subplot(224);
plt.stem(np.abs(fft(x2)));plt.axis([-1, 16, -0.5, 15]);
plt.xlabel('k');plt.ylabel('|X(k)|');
plt.title('16-point FFT of d)');

plt.suptitle('Reduction of the leakage effect by window functions: (a) the original signal,\n (b) the Blackman window function of length N = 8, \n'
            r'(c) product $x(n)\cdot wB(n) $ with $0 \leq n \leq N-1 $,(d) zero-padding applied to $x(n)\cdot wB(n)$' 
             '\n up to length N = 16 and the corresponding spectra are shown on the right side.',size=16)

plt.tight_layout()
plt.subplots_adjust(top=0.8)
plt.show()
```
