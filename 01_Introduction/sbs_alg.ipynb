```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.io import wavfile
import IPython
from IPython.display import Audio
%matplotlib inline
```

```python
# Read input sound file into vector x(n) and sampling frequency FS
AudioFileName = "ton2.wav"
Fs, X = wavfile.read(AudioFileName)
Audio(AudioFileName)
```

```python
# Ploting Signal
plt.rcParams['figure.figsize'] = (8, 3)# figure size
X = X*1.0/(2.**15) # Scale the signal in [-1, 1] for 16 bits input
plt.plot(X)
plt.title('Signal x(n)')
plt.ylabel('x(n)')
plt.xlabel('n')
plt.xlim([0,len(X)])
```

```python
# Sample-by sample algorithm y(n)=a*x(n)
# i multiply the signal for 2
a = 2
Y = a * X # each value of X is multiplied by a

# Write y(n) into output sound file with number of 
# bits Nbits and sampling frequency FS
wavfile.write('Output.wav',Fs,Y)
```

```python
# Ploting Signal
Fs, Y = wavfile.read('Output.wav')
#plt.rcParams['figure.figsize'] = (8, 3)# figure size
plt.subplot(121)
plt.plot(X)
plt.title('Signal x(n)')
plt.ylabel('x(n)')
plt.xlabel('n')
plt.xlim([0,len(X)])
plt.ylim([-1,1])
plt.subplot(122)
plt.plot(Y)
plt.title('Output Signal Y(n)')
plt.ylabel('Y(n)')
plt.xlabel('n')
plt.xlim([0,len(Y)])
plt.ylim([-1,1])
plt.tight_layout()
plt.show()
```
