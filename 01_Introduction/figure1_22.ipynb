```python
import matplotlib.pyplot as plt
from scipy import signal
from scipy.fftpack import fft
from scipy.io import wavfile
import IPython
from IPython.display import Audio
from waterfspec import waterfspec
%matplotlib inline
plt.style.use('ggplot')
```

```python
AudioFileName = "ton2.wav"
# Load audio File
Fs, data = wavfile.read(AudioFileName)
IPython.display.Audio(AudioFileName)
```

```python
# Ploting Signals
plt.rcParams['figure.figsize'] = (8, 3)# figure size
data = data*1.0/(2.**15) # Scale the signal in [-1, 1] for 16 bits input
plt.plot(data)
plt.title('Signal x(n)')
plt.ylabel('x(n)')
plt.xlabel('n')
plt.xlim([0,len(data)]);
```

```python
#waterfall spectrum plot
plt.rcParams['figure.figsize'] = (10, 5)# figure size
waterfspec(data,256,256,512,Fs,20,-100)
```

```python
N = 512
plt.rcParams['figure.figsize'] = (10, 4)# figure size
Pxx, freqs, bins, im = plt.specgram(data, NFFT=N, Fs=Fs,window = signal.blackman(N),noverlap = 0)
plt.title('specgram')
plt.ylabel('Freq in Hz')
plt.axis('tight');
```
